```python
import pandas as pd
```

```python
#getting relevent columns of content table
```

```python
content=pd.read_csv('content.csv')
```

```python
content.head()
```

```python
content.isnull().sum()
```

```python
content_category_typecontent=content[['Content ID' ,'User ID'  , 'Type' ,  'Category']]
content_category_typecontent.head()
```

```python
#getting relevent columns of reactions table
```

```python
reactions=pd.read_csv('reactions.csv')
```

```python
reactions.head()
```

```python
reactions.isnull().sum()
```

```python
reactions.shape
(980/25553)*100
```

##NAN values are 3% of data so we can drop the rows that don't have a type

```python
content_typereaction=reactions[['Content ID' , 'Type' , 'Datetime']]
content_typereaction.head()
```

```python
content_typereaction=content_typereaction.dropna()
content_typereaction.isna().sum()
```

```python
#getting relevent columns of ReactionTypes table
```

```python
ReactionTypes=pd.read_csv('ReactionTypes.csv')
```

```python
ReactionTypes.head()
```

```python
reaction_score=ReactionTypes[['Type' , 'Sentiment' , 'Score']]
reaction_score.head()
```

## Now our datasets are ready for merging 

```python
content_type_score=content_typereaction.merge(reaction_score , on='Type' )
content_type_score.head()
```

```python
content_type_score
```

```python
final_data=content_category_typecontent.merge(content_type_score , on='Content ID' , suffixes=['_content' , '_reaction'])
final_data.head()
```

```python
final_data.isnull().sum()
final_data.shape
```

```python
final_data.groupby('Category').agg('sum' , 'Score').sort_values('Score' , ascending=False)
```

```python
final_data['Category'].value_counts()
```

## searching the categories we find erroneous values like "dogs" is normally dogs and "food" and Food that must be replaced by food 
## so here we should clean our category columns from this erroneous values because it's gonna affect the final insights

## we have just 16 categories but there is repeated ones

```python
final_data['Category'] = final_data['Category'].replace("Fitness"  , "fitness").replace(['"soccer"' ,'Soccer'] , 'soccer').replace(['"science"' ,'Science'] , 'science').replace(['"animals"' ,'Animals'] , 'animals').replace(['"food"' ,'Food'] , 'food').replace(['"culture"' ,'Culture'] , 'culture').replace(['"studying"' ,'Studying'] , 'studying').replace(['"public speaking"' ,'Public Speaking'] , 'public speaking').replace(['"technology"' ,'Technology'] , 'technology').replace(['"veganism"' ,'Veganism'] , 'veganism').replace('"dogs"','dogs').replace('"cooking"' , 'cooking').replace('Travel' , 'travel').replace('Education' , 'education').replace('Healthy Eating' , 'healthy eating')
```

```python
final_data['Category'].value_counts()
```

### so here the top 5 categories are : animals / science / healthy eating / food / technology

```python
final_data['Category'].value_counts().head(5)
```

```python
final_data.to_csv('final_data.csv')
```
